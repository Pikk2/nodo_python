# Matplotlib

다양한 형태의 그래프를 통해서 데이터 시각화를 할 수 있는 라이브러리

```python
import matplotlib.pyplot as plt
```

# 1. 그래프 기본

```python
x = [2,4,8,16,32,64,128,256,512,1024]
y = [2,8,32,64,128,256,512,756,912,1024]

plt.plot(x,y)
plt.show()
```

## Title 설정

```python
plt.plot(x,y)
plt.title("Line Graph")
```

```python
plt.plot(x,y)
plt.title("꺾은선 그래프")
```
