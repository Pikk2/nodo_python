# 3. 범례(legend)

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"    # HYGung

matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 깨지는 현상 개선
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y, label = "Gamma 0.45")
plt.legend()
```

```python
plt.plot(x, y, label = "Gamma 0.45")
plt.legend(loc="upper left")
plt.legend(loc=(0.2, 0.8))  # 0~1 사이의 값
```
