# 4. 스타일

```python
import matplotlib.pyplot as plt
import matplotlib
matplotlib.rcParams['font.family'] = "Malgun Gothic"    # HYGung

matplotlib.rcParams['font.size'] = 15
matplotlib.rcParams['axes.unicode_minus'] = False # 마이너스 깨지는 현상 개선
```

```python
x = [1, 2, 3]
y = [2, 4, 8]
```

```python
plt.plot(x, y)
```

```python
plt.plot(x, y, linewidth=5)
```

```python
plt.plot(x,y, marker='o')
```

```python
plt.plot(x,y, marker='o', linestyle='None')
```

```python
plt.plot(x,y, marker='v', markersize=10)
```

```python
plt.plot(x,y, marker='X', markersize=10)
```

```python
plt.plot(x,y, marker='o', markersize=5, markeredgecolor='red')
```

```python
plt.plot(x,y, marker='o', markersize=15, markeredgecolor='red', markerfacecolor='yellow')
```

## 선 스타일

```python
plt.plot(x,y, linestyle = ":")
```

```python
plt.plot(x,y, linestyle = "--")
```

```python
plt.plot(x,y, linestyle = "-")
```

## 색깔

```python
plt.plot(x,y,color='pink')
```

```python
plt.plot(x,y,color='#FF0000',)
```

```python
plt.plot(x,y,color='b')
```

## 포맷

```python
plt.plot(x,y,'ro--') # color, marker, linestyle
```

```python
plt.plot(x,y,'bv:')
```

```python
plt.plot(x,y,'go') # color, marker, linestyle
```

## 축약어

```python
plt.plot(x,y, marker = 'o', mfc ='red', ms = 10, mec = 'blue', ls=':')
```

## 투명도

```python
plt.plot(x,y, marker = 'o', mfc ='red', ms = 10, alpha = 0.4)
```

## 그래프의 크기

```python
plt.figure(figsize=(5,10))
plt.plot(x,y)
```

```python
plt.figure(figsize=(10,5))
plt.plot(x,y)
```

```python
plt.figure(figsize=(2,2), dpi=200)
plt.plot(x,y)
```

## 배경색

```python
plt.figure(facecolor = "yellow",)
plt.plot(x,y)
```

```python
plt.figure(facecolor = "#a1c3ff",)
plt.plot(x,y)
```
